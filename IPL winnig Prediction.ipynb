# Data Preparation

```python
# There are many teams which can have diff diff names according to the ipl session. Like deccan charges is same as now sunrises hydrabad, so we are gonna replace this , and remove some times like pune which is not there in ipl.
match_df['team1'].unique()
```

```python
# Make a separrate variable of name of teams
teams=['Sunrisers Hyderabad', 'Mumbai Indians',
    'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Delhi Capitals', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals']
```

```python
match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team2']=match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')

```

```python
match_df=match_df[match_df['team1'].isin(teams)]
```

```python
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
match_df
```

```python
# Now take only those matches which has dl=0
match_df['dl_applied'].value_counts()
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
# Now we select the columns that is required in final prediction
match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
# because I want chasing team in my prediction so I will take only that rows which has inning =2
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.shape
```

```python
# columns that I want is batting_team, bowling_team , city,runs left , balls_left , wicket left , total _runs_x,rr,crr,result
```

```python
#Now we have to find runs left after every ball so 
delivery_df['current_score']=delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left']=126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
# for wickets_left
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna('0')
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x =='0' else '1')
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype('int')
wicket_left=delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets']=10 - wicket_left
delivery_df.head()
```

```python
# crr=runs/over
delivery_df['crr']=(delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr']=(delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df.head()
```

```python
# result column 
# if batting team == winner
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
# shuffle oit to avoid bias 
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

# Model Building

```python
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=1)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
